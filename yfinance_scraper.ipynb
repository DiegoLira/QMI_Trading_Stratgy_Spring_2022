```python
import pandas as pd
import yfinance as yf
import datetime
import time
import requests
from tqdm import tqdm
```

```python
'''Creating our Stock List of all Stocks in S&P 500'''

site = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
table = site[0]

sp_data = table
tickers = sp_data['Symbol'].values.tolist()
```

```python
'''Initial data table'''

stock_data = yf.download(tickers, start="2012-01-01", end="2022-04-01")

Adj_Close = stock_data['Adj Close'].T
Close = stock_data['Close'].T
Open = stock_data['Open'].T
Volume = stock_data['Volume'].T
Low = stock_data['Low'].T
High = stock_data['High'].T

stock_data
```

```python
for i in tqdm(range(0, len(stock_data.index))):
    day = stock_data.iloc[i]
    date = day.name
    date = str(date)[:-8]
    unstacked = day.unstack(1).T
    unstacked.to_csv(f'Data/{date}.csv')
```
